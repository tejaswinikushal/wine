#### Q2. How did you handle missing data in the wine quality data set during the feature engineering process? Discuss the advantages and disadvantages of different imputation techniques.

##### Mean/Median Imputation:

##### Advantages:
Simple and quick to implement.
Does not introduce bias to the dataset.
##### Disadvantages:
Ignores relationships between features.
May not be suitable for datasets with high variability.
##### Mode Imputation:

##### Advantages:
Suitable for categorical variables.
Preserves the original distribution of the data.
##### Disadvantages:
May not work well for continuous variables.
Ignores relationships between features.

##### Linear Regression Imputation:

##### Advantages:
Takes into account relationships between variables.
Provides more accurate estimates when relationships are linear.
#### Disadvantages:
Assumes a linear relationship, which may not be true.
Sensitive to outliers.

##### Considerations:
##### Nature of the Data:

Choose imputation methods based on the nature of the dataset (e.g., time-series, cross-sectional) and the type of features (e.g., numerical, categorical).
##### Amount of Missing Data:

For a small percentage of missing data, simple imputation methods like mean or median may be sufficient. For a larger percentage, more sophisticated methods might be necessary.
##### Domain Knowledge:

Consider the domain knowledge to choose an imputation method that makes sense for the specific context.

##### Q5. Load the wine quality data set and perform exploratory data analysis (EDA) to identify the distribution of each feature. Which feature(s) exhibit non-normality, and what transformations could be applied to these features to improve normality?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv"
df = pd.read_csv(url, sep=';')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.hist(bins=20, figsize=(15, 10))
plt.show()
```

```python
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
df['log_citric_acid'] = np.log1p(df['citric acid'])
```

```python
# Visualize the transformed distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['log_citric_acid'], bins=20, kde=True)
plt.title('Transformed Distribution of Log(Citric Acid Content)')
plt.xlabel('Log(Citric Acid Content)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Check for normality after transformation using a Q-Q plot
from scipy import stats
plt.figure(figsize=(10, 6))
stats.probplot(df['log_citric_acid'], plot=plt)
plt.title('Q-Q Plot for Transformed Citric Acid Content')
plt.show()
```
